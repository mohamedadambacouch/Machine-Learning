#### I)1)a) Extracting the Standings : 

```python
def get_standings_df(url):
    # Send a request to the URL
    data = requests.get(url)
    # Parse the content using BeautifulSoup
    soup = BeautifulSoup(data.content, 'html.parser')
    # Find all the table rows
    table_rows = soup.find_all('tr')
    
    rank = [row.find_all('td')[0].get_text().strip() for row in table_rows[1:21]]
    teams = [row.find_all('td')[1].get_text().strip() for row in table_rows[1:21]]
    points = [row.find_all('td')[2].get_text().strip() for row in table_rows[1:21]]
    played = [row.find_all('td')[3].get_text().strip() for row in table_rows[1:21]]
    goal_diff = [row.find_all('td')[4].get_text().strip() for row in table_rows[1:21]]
    wins = [row.find_all('td')[5].get_text().strip() for row in table_rows[1:21]]
    draws = [row.find_all('td')[6].get_text().strip() for row in table_rows[1:21]]
    losses = [row.find_all('td')[7].get_text().strip() for row in table_rows[1:21]]
    goals_for = [row.find_all('td')[8].get_text().strip() for row in table_rows[1:21]]
    goals_against = [row.find_all('td')[9].get_text().strip() for row in table_rows[1:21]]
    
    # Create a DataFrame from the extracted data
    df = pd.DataFrame({
        'Rank': rank,
        'Team': teams,
        'Played': played,
        'Wins': wins,
        'Draws': draws,
        'Losses': losses,
        'Goals For': goals_for,
        'Goals Against': goals_against,
        'Goal Difference': goal_diff,
        'Points': points
    })
    
    return df
```

```python
def get_stats(url):
    # Send a request to the URL
    data = requests.get(url)
    # Parse the content using BeautifulSoup
    soup = BeautifulSoup(data.content, 'html.parser')
    # Find all the table rows
    table_rows = soup.find_all('tr')
    
    teams = [row.find_all('td')[0].get_text().strip() for row in table_rows[1:21]]
    Num_players = [row.find_all('td')[1].get_text().strip() for row in table_rows[1:21]]
    Num_foreigners = [row.find_all('td')[2].get_text().strip() for row in table_rows[1:21]]
    Avg_age = [row.find_all('td')[3].get_text().strip() for row in table_rows[1:21]]
    Salary = [row.find_all('td')[4].get_text().strip() for row in table_rows[1:21]]
    Expenses = [row.find_all('td')[5].get_text().strip() for row in table_rows[1:21]]
    Avg_Player_Rate = [row.find_all('td')[6].get_text().strip() for row in table_rows[1:21]]
    
    # Create a DataFrame from the extracted data
    df = pd.DataFrame({
        'Team': teams,
        'Num players': Num_players,
        'Num foreigners': Num_foreigners,
        'Avg age': Avg_age,
        'Salary': Salary,
        'Expenses': Expenses,
        'Avg Player Rate': Avg_Player_Rate
    })
    
    return df
```

### II)2) Feature Enginnering : 

```python
def convert_money_values(value):
    # Replace commas with periods
    value = value.replace(',', '.')
    # Convert 'M' to millions and 'K' to thousands
    if 'M' in value:
        return float(value.replace('M', ''))
    elif 'K' in value:
        return float(value.replace('K', '')) / 1000
    else:
        return float(value)
```

```python

# Apply this function to the Salary and Expenses columns of the France dataframe
France['Salary'] = France['Salary'].apply(convert_money_values)
France['Expenses'] = France['Expenses'].apply(convert_money_values)
# Apply to England
England['Salary'] = England['Salary'].apply(convert_money_values)
England['Expenses'] = England['Expenses'].apply(convert_money_values)

# Apply to Germany
Germany['Salary'] = Germany['Salary'].apply(convert_money_values)
Germany['Expenses'] = Germany['Expenses'].apply(convert_money_values)

# Apply to Italy
Italy['Salary'] = Italy['Salary'].apply(convert_money_values)
Italy['Expenses'] = Italy['Expenses'].apply(convert_money_values)

# Apply to Spain
Spain['Salary'] = Spain['Salary'].apply(convert_money_values)
Spain['Expenses'] = Spain['Expenses'].apply(convert_money_values)
```

```python
def evaluate_random_forest(data):
    # Drop non-feature columns
    df=data.copy()
    df.drop(columns=['Rank', 'Team'], inplace=True)

    # Prepare data
    X = df.drop('Points', axis=1)
    y = df['Points']
    df_test = df[df['Year'] == 2024]
    X_test = df_test.drop(columns=['Points', 'Year'])
    y_test = df_test['Points']

    df_train = df[df['Year'] < 2024]
    X_train = df_train.drop(columns=['Points', 'Year'])
    y_train = df_train['Points']

    # Train Random Forest Classifier
    rf_classifier = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
    rf_classifier.fit(X_train, y_train)
    y_pred_classifier = rf_classifier.predict(X_test)

    # Train Random Forest Regressor
    rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_regressor.fit(X_train, y_train)
    y_pred_regressor = rf_regressor.predict(X_test)

    # Calculate evaluation metrics
    mae = mean_absolute_error(y_test, y_pred_regressor)
    r2 = r2_score(y_test, y_pred_regressor)

    return mae, r2 
```

```python
# Evaluate the model for each country
Fr_mae, Fr_r2 = evaluate_random_forest(France)
Ger_mae, Ger_r2 = evaluate_random_forest(Germany)
Sp_mae, Sp_r2 = evaluate_random_forest(Spain)
Ita_mae, Ita_r2 = evaluate_random_forest(Italy)
Eng_mae, Eng_r2 = evaluate_random_forest(England)

# Create a dictionary to hold the results
results = {
    'France': [Fr_mae, Fr_r2],
    'Germany': [Ger_mae, Ger_r2],
    'Spain': [Sp_mae, Sp_r2],
    'Italy': [Ita_mae, Ita_r2],
    'England': [Eng_mae, Eng_r2]
}

# Convert the dictionary to a DataFrame
results_df = pd.DataFrame(results, index=['MAE', 'R_squared'])
```

```python
import pandas as pd

def create_league_metrics(df):
    # Calculate Goals For per Game
    df['GF/Game'] = df['Goals For'] / df['Played']

    # Calculate Goals Against per Game
    df['GA/Game'] = df['Goals Against'] / df['Played']

    # Calculate Goal Difference per Game
    df['GD/Game'] = df['Goal Difference'] / df['Played']
    return df

# Apply the function to each dataframe
France = create_league_metrics(France)
England = create_league_metrics(England)
Germany = create_league_metrics(Germany)
Italy = create_league_metrics(Italy)
Spain = create_league_metrics(Spain)
```

```python
def calculate_averages(df):
    avg_gf_game = df['GF/Game'].mean()
    avg_ga_game = df['GA/Game'].mean()
    avg_gd_game = df['GD/Game'].mean()
    return avg_gf_game, avg_ga_game, avg_gd_game

# Calculate averages for each country
avg_metrics = {}
avg_metrics['France'] = calculate_averages(France)
avg_metrics['England'] = calculate_averages(England)
avg_metrics['Germany'] = calculate_averages(Germany)
avg_metrics['Italy'] = calculate_averages(Italy)
avg_metrics['Spain'] = calculate_averages(Spain)

# Create a DataFrame to display the results
avg_metrics_df = pd.DataFrame(avg_metrics, index=['Avg GF/Game', 'Avg GA/Game', 'Avg GD/Game'])

avg_metrics_df
```

```python
import matplotlib.pyplot as plt

# Assuming avg_metrics_df is already defined as described previously

# Plotting the averages
plt.figure(figsize=(12, 6))

# Bar plot for Avg GF/Game
plt.subplot(1, 3, 1)
plt.bar(avg_metrics_df.columns, avg_metrics_df.loc['Avg GF/Game'], color='b')
plt.xlabel('Country')
plt.ylabel('Avg GF/Game')
plt.title('Average Goals For per Game')

# Bar plot for Avg GA/Game
plt.subplot(1, 3, 2)
plt.bar(avg_metrics_df.columns, avg_metrics_df.loc['Avg GA/Game'], color='r')
plt.xlabel('Country')
plt.ylabel('Avg GA/Game')
plt.title('Average Goals Against per Game')

# Bar plot for Avg GD/Game
plt.subplot(1, 3, 3)
plt.bar(avg_metrics_df.columns, avg_metrics_df.loc['Avg GD/Game'], color='g')
plt.xlabel('Country')
plt.ylabel('Avg GD/Game')
plt.title('Average Goal Difference per Game')

plt.tight_layout()
plt.show()
```

### After it we can observe that England and Italy snatch the second place when it comes to Goals per game , and finely France and spain are the least scoring leagues. 

### IV)1)b) Number Of Competative Teams: 

```python
def count_teams_with_high_win_ratio(df, country_name):
    df['Win_Ratio'] = pd.to_numeric(df['Win_Ratio'], errors='coerce')
    count_per_year = df.groupby('Year').apply(lambda x: (x['Win_Ratio'] >= 0.5).sum()).reset_index(name='High_Win_Ratio_Count')
    print(f"Number of teams in {country_name} with Win_Ratio >= 0.5 grouped by Year:\n{count_per_year}")
    return count_per_year
```

```python
france_high_win_ratio_count = count_teams_with_high_win_ratio(France, 'France')
england_high_win_ratio_count = count_teams_with_high_win_ratio(England, 'England')
germany_high_win_ratio_count = count_teams_with_high_win_ratio(Germany, 'Germany')
italy_high_win_ratio_count = count_teams_with_high_win_ratio(Italy, 'Italy')
spain_high_win_ratio_count = count_teams_with_high_win_ratio(Spain, 'Spain')
```

```python
# Create dataframes for each country's high win ratio count
df_france = france_high_win_ratio_count.set_index('Year').rename(columns={'High_Win_Ratio_Count': 'France'})
df_england = england_high_win_ratio_count.set_index('Year').rename(columns={'High_Win_Ratio_Count': 'England'})
df_germany = germany_high_win_ratio_count.set_index('Year').rename(columns={'High_Win_Ratio_Count': 'Germany'})
df_italy = italy_high_win_ratio_count.set_index('Year').rename(columns={'High_Win_Ratio_Count': 'Italy'})
df_spain = spain_high_win_ratio_count.set_index('Year').rename(columns={'High_Win_Ratio_Count': 'Spain'})

# Concatenate dataframes along the columns
combined_df = pd.concat([df_france, df_england, df_germany, df_italy, df_spain], axis=1)
```

```python
# Calculate the average for each country
avg_by_country = combined_df.mean(axis=0)

# Print the average for each country
print("Average by Country:")
avg_by_country
```
